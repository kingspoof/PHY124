```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import transforms
from scipy.ndimage import rotate

filename = 'data/holbein.png'
```

```python
def cut_image(image, start_width, start_height, width, height):
    """
    Cuts an image into a smaller image
    """
    return image[start_height:start_height+height, start_width:start_width+width]

def rotate_image(image, degree):
    """
    Rotates an image 90, 180 or 270 degress
    degrees is 1 = 90, 2 = 180, ...
    """
    return rotate(image, degree)

def stretch(image, width_index = 1, height_index = 1):
    ret = 0*image

    width = image.shape[0]
    height = image.shape[1]

    width_half = width//2
    height_half = height//2

    for w in range(width):
        for h in range(height):

            dw = w - width_half
            dh = h - height_half

            sw = width_half + dw * width_index
            sh = height_half + dh * height_index
            if 0 <= sw < width and 0 <= sh < height:
                ret[w, h, :] = image[int(sw), int(sh), :]
    
    return ret
    
```

```python

filename = 'data/holbein.png'



pic = plt.imread(filename)
#print(len(pic), len(pic[0]))
#print(pic[0][0])

pic = cut_image(pic, 180, 770, 650, 260)
pic = rotate_image(pic, -25)
pic = stretch(pic, 0.2, 2)
plt.imshow(pic)
```
