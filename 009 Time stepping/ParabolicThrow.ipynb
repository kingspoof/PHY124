```python
def add(a,b):
    if(len(a) != len(b)):
        ret = [0 for n in a]
        for i in len(a):
            ret[i] = a[i] + b[i]
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def a(r, t):
    return [0, -9.81]


r = [[0,0]]
v = [[14,5]]
t = [0]

dt = 0.4
while t[-1] < (5 - dt/2):
    # first step
    t.append(t[-1] + dt/2)
    r.append([r[-1][0] + v[-1][0] * dt/2, r[-1][1] + v[-1][1] * dt/2])

        
    v.append([v[-1][0] + a(r[-1], t[-1])[0] * dt ,v[-1][1] + a(r[-1], t[-1])[1] * dt])

    t.append(t[-1] + dt/2)
    r.append([r[-1][0] + v[-1][0] * dt/2, r[-1][1] + v[-1][1] * dt/2])
```
