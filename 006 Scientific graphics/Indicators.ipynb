```python
import csv
import math
import numpy as np
from matplotlib import pyplot as plt

selected_countires = ['AZE', 'KAZ', 'GEO', 'BLR', 'LTU', 'EST', 'UKR']

filename = 'data/War.csv'
with open(filename, 'r') as file:
    reader = csv.reader(file)
    data = list(reader)


year_start = 1960 # this is valid from 1960
year_end = 2022 # this is valid to 2022



header = [float(val) for val in data[4][(year_start- 1955):(year_end - 1955)]]
countries = {}

for value in data[4:]:
    if(value[1] in selected_countires):
        countries[value[1]] = {
            'name': value[0],
            'short': value[1],
            'spending': [float(val) if val != '' else np.nan for val in value[(year_start- 1955):(year_end - 1955)]] #value[45:67]
        }

# instead of 0 we can use np.nan

for cunt in countries:
    plt.plot(header, countries[cunt]['spending'], label=countries[cunt]['name'])

plt.title('Military spending of countries bordering russia')

#plt.axvline(x=2022, label='Invasion of Ukraine')

plt.xlabel('Year')
plt.ylabel('Military spending as % of GDP')
plt.legend()


```
