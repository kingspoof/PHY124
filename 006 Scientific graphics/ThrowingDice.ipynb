```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# a space to define functions
def sum_values(dice_line):
    sum = 0
    for value in dice_line.rstrip().split(' '):
        sum += int(value)
    return sum
```

```python
filename = 'data/dice.txt'
dice_file = open(filename, 'r')

# the file as is and the summed values
dice_info = [line for line in dice_file.readlines()]
dice_summed = [sum_values(entry) for entry in dice_info]
```

```python
plt.plot(dice_summed[:100], '--bo', markersize=2)
plt.title('first 100 repetitions')
plt.xlabel('Number of repetition')
plt.ylabel('Summed value')
#plt.legend()
```

```python
data = np.array(dice_summed)
y, bin_edges = np.histogram(data, bins=range(3,20))
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])
mean_standard_deviation = np.sqrt(y)
width = 1

plt.bar(
    bin_centers,
    y,
    width=width,
    color='lightblue',
    edgecolor='white'
)
plt.errorbar(
    bin_centers,
    y,
    yerr=mean_standard_deviation,
    xerr=0.5,
    fmt='none',
    ecolor='red',
)


plt.xlabel('Value')
plt.ylabel('Number of repetitions')
```
